# Importing libraries

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
```

# Importing data

```python
data=pd.read_csv("D:\\Workshops\\Machine Learning for Data Science & Artifcial Intelligence With Python\\data\\movie_data.CSV")
```

```python
data.head()
```

# Data wrangling

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data=data[~data["movie_title"].duplicated()]
```

```python
data.reset_index(inplace=True)
```

```python
data.head()
```

```python
data.drop("index",axis=1,inplace=True)
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

# Needful columns

```python
df=data[["movie_title","director_name","actor_1_name","actor_2_name","genres","country","language"]]
```

```python
df.head()
```

# Combined features

```python
" ".join(list(df.iloc[0].values))
```

```python
features=[]

for i in range(df.shape[0]):
    features.append(" ".join(list(df.iloc[i].values)))
```

```python
df["features"]=features
```

# Fetaure count matrix

```python
cvec=CountVectorizer()
```

```python
cv_df=cvec.fit_transform(df["features"])
```

```python
cv_df.toarray()
```

```python
cv_df.shape
```

```python
df.shape
```

# Cosine similarity

```python
cs=cosine_similarity(cv_df)
```

```python
cs
```

```python
cs.shape
```

# Recommendations

```python
title="Avatar"
movie_index=df[[title in name for name in df["movie_title"]]].index[0]
```

```python
movie_index
```

```python
scores=list(enumerate(cs[movie_index]))
```

```python
scores
```

```python
sorted_scores=sorted(scores,key=lambda x:x[1],reverse=True)
```

```python
sorted_scores
```

```python
top_10=[h[0] for h in sorted_scores[1:11]]
```

```python
top_10
```
