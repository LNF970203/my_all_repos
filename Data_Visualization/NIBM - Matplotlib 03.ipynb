## Drawing shapes in Matplotlib

- As best Python package for drawing plots, Matplotlib also has impressive primitive drawing capablities

## Drawing  a circle

[Circle other parameters](https://matplotlib.org/stable/api/_as_gen/matplotlib.patches.Circle.html)

```python
import matplotlib.pyplot as plt

plt.axes()

circle = plt.Circle((0, 0), radius=5, fc='y', ec='g')
plt.gca().add_patch(circle)

plt.axis('scaled') ## make sure that entire plot is visible
plt.show()
```

## Drawing  a Rectangle

[Rectangle other parameters](https://matplotlib.org/stable/api/_as_gen/matplotlib.patches.Rectangle.html#matplotlib.patches.Rectangle)

```python
rectangle = plt.Rectangle((50, 10), 100, 50, fc='r')
plt.gca().add_patch(rectangle)
plt.axis('scaled') ## make sure that entire plot is visible
plt.show()
```

```python
rectangle = plt.Rectangle((10, 10), 10, 20, fill=True,ec='g')
rectangle2 = plt.Rectangle((18, 25), 25, 15, fill=False,ec='g')
plt.gca().add_patch(rectangle)
plt.gca().add_patch(rectangle2)
plt.axis('scaled') 
plt.show()
```

## Drawing Lines

[Lines other parameters](https://matplotlib.org/stable/api/_as_gen/matplotlib.lines.Line2D.html)

```python
line = plt.Line2D((20, 30), (10, 15), lw=2.5)
plt.gca().add_line(line)
plt.axis('scaled') 
plt.show()
```

```python
dotted_line = plt.Line2D((2, 8), (4, 5), lw=5., 
                         ls='-.', marker='*', 
                         markersize=20, 
                         markerfacecolor='r', 
                         markeredgecolor='g', 
                         alpha=1)
plt.gca().add_line(dotted_line)
plt.axis('scaled') 
plt.show()
```

## Drawing Polygons

[Polygons other parameters](https://matplotlib.org/stable/api/_as_gen/matplotlib.patches.Polygon.html#matplotlib.patches.Polygon)

```python
points = [[2, 4], [2, 8], [4, 6], [6, 8], [6,4]]
line = plt.Polygon(points, closed=None, fill=None, edgecolor='r')
plt.gca().add_line(line)
plt.axis('scaled') 
plt.show()
```

```python
points = [[2, 4], [2, 8], [4, 6], [6, 8], [6,4]]
line = plt.Polygon(points, closed=True, fill=True, edgecolor='r')
plt.gca().add_line(line)
plt.axis('scaled') 
plt.show()
```

```python
arr = plt.Arrow(4,5,2,3, width=2)
plt.gca().add_patch(arr)
plt.axis('scaled') 
plt.show()
```

### Text inside a matplotlib plot.

- we can annotate  annotate certain features of the plot to give a critical information. 

- This can be done manually with the __plt.text/ax.text__ command, which will place text at a particular x/y value

- we can style the font using a dictionary



For more info on Annotation visit [here](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.text.html)

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

slcd = pd.read_csv(r"C:\Users\Sanjeth\Desktop\NIBM lectures\DATA VISUALIZATION AND INTERACTIVE DASHBOARDS USING PYTHON\Full time\Matplotlib lec 4 5 6\lec4\srilanka covid data.csv")

fig, ax = plt.subplots()  # Create a figure and an axes.
ax.plot(slcd['time line'], slcd['new_cases'], label='New Cases')


# add pathch 1
arr = plt.Arrow(150,500,100,-500, width=20, color = 'r')
ax.add_patch(arr)


# add patch 2 
rectangle = plt.Rectangle((360, 200), 90, 500, fill=False,ec='g')
ax.add_patch(rectangle)

# Add labels to the plot
style = dict(size=10, color='black')
ax.text(150, 200, "1st wave start", **style, rotation=320)
ax.text(360, 50, "Lock Down", **style)

plt.show()
```

### Three-Dimensional Plotting in Matplotlib
 - Matplotlib was initially designed with only two-dimensional plotting in mind. 
 - Around the time of the 1.0 release, some three-dimensional plotting utilities were built on top of Matplotlib’s two-dimensional display, and the result is a convenient (if somewhat limited) set of tools for three-dimensional data visualization.
 - We enable three-dimensional plots by importing the **mplot3d toolkit**, included with the main Matplotlib installation
 - Once this submodule is imported, we can create a three-dimensional axes by passing the keyword projection='3d' to any of the normal axes creation routines:

```python
import matplotlib
matplotlib.__version__
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
```

- The most basic three-dimensional plot is a line or scatter plot created from sets of (x, y, z) triples.
- In analogy with the more common two-dimensional plots discussed earlier, we can create these using the ax.plot3D and ax.scatter3D functions
- The call signature for these is nearly identical to that of their two-dimensional counterparts, so you can refer to “Simple Line Plots” and “Simple Scatter Plots” for more information on controlling the output.

```python
ax = plt.axes(projection='3d')


zline = np.linspace(0, 15, 1000)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline, yline, zline, 'Blue')
```

```python
ax = plt.axes(projection='3d')
zdata = 25 * np.random.random(100)
xdata = np.sin(zdata) + 0.1 * np.random.randn(100)
ydata = np.cos(zdata) + 0.1 * np.random.randn(100)
p = ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='Blues');

fig.colorbar(p, ax=ax) ## mntion plot and colorbar
#fig.colorbar(ax)
```

## Three- Dimensional Bar plots

```python
import random
fig = plt.figure()
ax = plt.axes(projection="3d")

num_bars = 15
x_pos = random.sample(range(20), num_bars)
y_pos = random.sample(range(20), num_bars)
z_pos = [0] * num_bars
x_size = 2*np.ones(num_bars)
y_size = 3*np.ones(num_bars)
z_size = random.sample(range(20), num_bars)

ax.bar3d(x_pos, y_pos, z_pos, x_size, y_size, z_size, color='aqua', alpha=0.3)
plt.show()
```

### Application of using 3d barchart in Matplotlib

```python
import pandas as pd
path = r"C:\Users\Sanjeth\Desktop\NIBM lectures\Data visualisation with python - ADDS\Matplotlib Lec 5,6,7\lec5\owid-covid-data.csv"
df = pd.read_csv(path)
df.dtypes

loc_avg = df.groupby('location').agg({'total_cases':'mean'})
sorted_df = loc_avg.sort_values('total_cases',ascending=False)
top_5_contries = loc_avg.sort_values('total_cases',ascending=False)[4:9].index

df_top = df[df['location'].isin(list(top_5_contries))]
df_top.loc[:,'date'] = pd.to_datetime(df_top['date'])
df_top_21 = df_top[df_top['date']> '2021']
df_top_21.loc[:,'Month'] = df_top['date'].dt.month
df_top_21_filtered = df_top_21[['Month','location','new_cases']]



pivot_ = df_top_21_filtered.pivot_table(index='Month', columns='location', aggfunc={'new_cases':'mean'})
pivot_.columns = pivot_.columns.droplevel(0) 


rows_, columns_= pivot_.shape
xpos = np.arange(0, rows_, 1)
ypos = np.arange(0, columns_, 1)
xpos, ypos = np.meshgrid(xpos, ypos)

xpos = xpos.flatten()
ypos = ypos.flatten()
zpos = np.zeros(rows_*columns_)


width_, lenght_ = 0.25, 0.25
x_val = width_*np.ones(rows_*columns_)
y_val = lenght_*np.ones(rows_*columns_)
z_val = pivot_.values.flatten()

fig = plt.figure(figsize=(8, 10))
ax = plt.axes(projection="3d")

colors = ['red', 'aqua', 'blue', 'brown', 'green']*columns_
ax.bar3d(xpos, ypos, zpos, x_val, y_val, z_val,color=colors)
ax.w_xaxis.set_ticklabels(['0', 'Brazil', 'European Union', 'India', 'South America', 'United States'])
ax.w_yaxis.set_ticklabels([0,1, 2, 3, 4, 5])
plt.show()
```
