## Histogram

A histogram is used to summarize discrete or continuous data.  

It provides a visual interpretation. This requires focusing on the main points, facts of numerical data by showing the number of data points that fall within a specified range of values (called “bins”)


```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')

data = np.random.randn(1000)
```

```python
plt.hist(data)
```

## other  options

- alpha      - color darkness
- histype    - how the histogram should be filled [bar, barstacked, step, stepfilled]
- colour     - color of the filling
- edgecolor  - edge color
- align      - The horizontal alignment of the histogram bars
- cumulative - whether the graph is cumalative


```python
plt.hist(data, bins=10, alpha=0.6,
         histtype='stepfilled', color='green',
         edgecolor='blue', align = 'mid')
plt.title("Normal histogram");
```

```python
plt.hist(data, bins=10, alpha=0.6,
         histtype='stepfilled', color='green',
         edgecolor='blue', align = 'mid' ,cumulative = True)
plt.title("Cumulative Histogram");
```

### Specifying bins in histogram

- integer: number of equal sized bins
- sequence: bing edges including  first left and  last right edges (except last bin all are half open less than or equal)
- string: 'auto', 'fd', 'doane', 'scott', 'stone', 'rice', 'sturges', or 'sqrt'. [refer this url for different binning stratergies](https://www.answerminer.com/blog/binning-guide-ideal-histogram)


```python
plt.hist(data, bins= 'sqrt', alpha=0.6,
         histtype='stepfilled', color='green',
         edgecolor='blue', align = 'mid')
plt.title("Defining custom bing edges");
```

```python
plt.hist(data, bins=[-3,-2,0,2,3], alpha=0.6,
         histtype='stepfilled', color='green',
         edgecolor='blue', align = 'mid')
plt.title("Defining custom bing edges");
```

## density plot

# $ density = \frac{counts}{total\,\#\,of\,points\,\,*\,\,bin\,width}$

- so that the area under the histogram integrates to 1

```python
## density plot
plt.hist(data, bins=10, alpha=0.6,
         histtype='stepfilled', color='green',
         edgecolor='blue', align = 'mid', density=True)
plt.title("Defining custom bing edges");

```

```python
## density plot
plt.hist(data, bins=10, alpha=0.6,
         histtype='stepfilled', color='green',
         edgecolor='blue', align = 'mid', density=True, cumulative=True)
plt.title("Defining custom bing edges");

```

## getting bin and count as histogram in table


```python
counts, bin_edges = np.histogram(data, bins=10)
print(counts)
print(bin_edges)
```

## 2D Histogram

- Here distribution of 2 continous varibales are compared in the same graph.
- bin information of 2 variables will be specified in 2 axis and the count will be represented by darkness of the color 

```python
x1 = np.random.normal(3, 2, 1000)
x2 = np.random.normal(-2, 1, 1000)
```

```python
plt.hist2d(x1, x2, bins=10, cmap='Blues')
cb = plt.colorbar()
cb.set_label('counts of bin')
```

# getting bin and count as 2d histogram in table



```python
counts, xedges, yedges = np.histogram2d(x1, x2, bins=5)
print(counts)
print(xedges)
print(yedges)
```

## Hexbin

A hexbin plot is useful to represent the relationship of 2 numerical variables when you have a lot of data points. Without overlapping of the points, the plotting window is split into several hexbins. The color of each hexbin denotes the number of points in it.  

hexagons are very clear and desired over squares and thats why hexbin are quite famous than 2D histograms in matplotlib.

hexagons are natural shape used for tesselation


```python
plt.hexbin(x1, x2, gridsize=10, cmap='Blues')
cb = plt.colorbar(label='count of bin')
```

## other options

- gridsize = int (x direction and y is adjusted to regular) or (int,int) 
- c = to specify value counts so x anf y will be treated as labels
- bins = unlike histogram here the job of bin is discretising the hexagon value >> log, integer, sequence
- mincnt = if not None, only display cells with more than mincnt number of points in the cell.
- alpha, edgecolor, linewidth, cmap

```python
plt.hexbin(x1, x2, gridsize=(10,5), cmap='Blues')
cb = plt.colorbar(label='count of bin')
```

```python
# how 3rd array works
a = [1,2,3,4,5,6]
b = [3,4,5,6,7,8]
c = [34,56,76,45,44,56]
plt.hexbin(a, b, c,gridsize=10, cmap='Blues')
cb = plt.colorbar(label='count of bin')
```

```python
plt.hexbin(x1, x2, gridsize=10, cmap='Blues',bins=5 )
cb = plt.colorbar(label='levels of bin')
```

```python
plt.hexbin(x1, x2, gridsize=10, cmap='Blues',bins='log' )
cb = plt.colorbar(label='levels of bin')
```

```python
plt.hexbin(x1, x2, gridsize=10, cmap='Blues',bins=[1,2,4] )
cb = plt.colorbar(label='levels of bin')
```

## kernal density estimation

In statistics, kernel density estimation (KDE) is a non-parametric way to estimate the probability density function of a random variable. Kernel density estimation is a fundamental data smoothing problem where inferences about the population are made, based on a finite data sample.

```python
from scipy.stats import gaussian_kde

# fit an array of size [Ndim, Nsamples]
data = np.vstack([x1, x2])
kde = gaussian_kde(data)

print(kde)
# evaluate on a regular grid
xgrid = np.linspace(0, 6, 100)
ygrid = np.linspace(-4, 0, 100)
Xgrid, Ygrid = np.meshgrid(xgrid, ygrid)


Z = kde.evaluate(np.vstack([Xgrid.ravel(), Ygrid.ravel()]))

# Plot the result as an image
plt.imshow(Z.reshape(Xgrid.shape),
           origin='lower', aspect='auto',
           extent=[0, 6, -4, 0],
           cmap='Blues')
cb = plt.colorbar()
cb.set_label("density")
```

## pie charts

- pie charts used to show the counts of different factors of a categorigal varible.
- pie charts helps to compare the different factors of a varible.

### simple pie chart

```python

y = np.array([25, 17, 34, 23])

plt.pie(y)
plt.show()
```

### simle pie chart with labels

```python
y = np.array([25, 17, 34, 23])
lbl = ['A','B','C','D']
plt.pie(y, labels=lbl)
plt.show()
```

## starting angle 

- defalut starting angle is x axis =0
- angles specified will be applied on anticlockwise

```python
y = np.array([25, 17, 34, 23])
lbl = ['A','B','C','D']
plt.pie(y, labels=lbl, startangle = 180)
plt.show()
```

## explode some wedges

- this will help to get some wedges out of the pie chart to uniquely specify that factor
- can be given in the range of 0 to 1 >> 0 - fully attached to center and 1 - fully out of center

```python
y = np.array([25, 17, 34, 23])
lbl = ['A','B','C','D']
epld = [0.2,0,0.5,0]
plt.pie(y, labels=lbl, startangle = 180, explode = epld)
plt.show()
```

## shadow ,color, legend
- shadow can be enabled
- matplotlib accpeted colors can be specified 
- simply legend can be apllied if necessary and all the legend properties will work


```python
y = np.array([25, 17, 34, 23])
lbl = ['A','B','C','D']
epld = [0.4,0,0,0]
cls = ["black", "green", "r", "#BBAF50"]
plt.pie(y, labels=lbl, startangle = 180, explode = epld, colors = cls, shadow=True)
plt.legend(title = "Letters")
plt.show() 
plt.show()
```

## Bar charts 

A bar chart or bar graph is a chart or graph that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent. The bars can be plotted vertically or horizontally.


## Vertical Bar chart

- width value will define the widht of the bar
- bottom value will define the y cordinate of bar

```python

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y)
plt.show()
```

```python

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y, width = 0.3)
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y, width = 0.3, bottom=[2,6,7,8])
plt.show()
```

## Horizontal Bar chart

- height value will define the height of the bar
- left value will define the x cordinate of bar

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x,y)
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x,y, height= 0.3)
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x,y, height= 0.3, left=[2,6,7,4])
plt.show()
```

## step plots

A stepped line graph (also called step chart) is a chart similar to a line graph, but with the line forming a series of steps between data points. A stepped line chart can be useful when you want to show the changes that occur at irregular intervals.

```python
x = np.array([0, 1, 2, 3, 5])
y = np.array([3, 7, 9, 6, 13])
plt.step(x, y);
```

```python
x = np.array([0, 1, 2, 3, 5])
y = np.array([3, 7, 9, 6, 13])
plt.step(x, y, where='post');
plt.step(x, y, where='mid');
```
