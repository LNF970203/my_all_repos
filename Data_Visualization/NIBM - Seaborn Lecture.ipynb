## SEABORN

- Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

- Seaborn is a complement, not a substitute, for Matplotlib. Sometimes for advance visual editing we might need matploltib  


### Why Seaborn over matplotlib:
    1. Making attractive statistical plots.
    2. Easily and flexibly displaying distributions.
    3. Visualizing information from matrices and DataFrames.

### Seaborn setting Context

-  Seaborn also allows us to control individual elements of our graphs and thus we can control the scale of these elements or the plot by using the set_context() function
    
 
  
- This affects things like the size of the labels, lines, and other elements of the plot, but not the overall style. The base context is “notebook”, and the other contexts are “paper”, “talk”, and “poster”, which are version of the notebook parameters scaled by .8, 1.3, and 1.6, respectively.

#### KDE plot

- A kernel density estimate (KDE) plot is a method for visualizing the distribution of observations in a dataset, analagous to a histogram. KDE represents the data using a continuous probability density curve in one or more dimensions.

- Kernel density estimation is a fundamental data smoothing problem where inferences about the population are made, based on a finite data sample

```python
sns.kdeplot(sl_df['total_cases'])
```

#### ECDF plot

- Empirical Commutative Distribution
- ECDF plot visualizes each and every data point of the dataset directly, without binsize setting.
- x-axis corresponds to the range of values for the variable whereas the y-axis corresponds to the proportion of data points that are less than or equal to the corresponding value of the x-axis.

```python
sns.ecdfplot(x = sl_df['total_cases'])
```

## Plots with Categorical variable.

### 3 types of categorical plots  available. (with a numerical variable)

1. Plotting Categorical Estimate Plots
2. Plotting Categorical Distribution plots
3. Plotting Categorical Scatter plots


## 1. Plotting Categorical Estimate Plots  
##### Plots which show an estimate of the central tendency of the values
- Bar plot
- count plot
- point plot

#### 1.1 Bar plot
- It operates on a full dataset and applies a function to obtain the estimate 
- It can accept long data frame as well as wide data frame and also pandas series, numpy series and lists

```python
sns.barplot(x = 'location', y = 'new_cases', data = w_df,
            palette = 'cividis', ci = 95 )
```

- Additionaly bar plot uses bootstrapping to compute a confidence interval for the estimate 

```python
sns.barplot(x = 'location', y = 'total_cases', data = w_df,
            palette = 'cividis', ci = 5 )
```

```python
from numpy import count_nonzero
md = w_df.iloc[:20]
ax = sns.barplot(x = 'new_deaths', y = 'total_cases', data = md, estimator= count_nonzero)
```

```python
sns.barplot(y = 'location', x = 'total_cases', data = w_df, hue = 'Year', orient = 'h')
```

```python
sns.barplot(x = 'location', y = 'total_cases', data = w_df, hue = 'Year')
```

```python
import numpy as np
```

```python
get_median = lambda x: np.percentile(x, 50)

sns.barplot(y = 'location', x = 'new_cases', data = w_df, orient = 'h', estimator = get_median)
```

```python
sns.barplot(y = ['b','b','c','c','d','e'], x = [2,4,5,4,5,6], data = w_df, orient = 'h', estimator = lambda x: len(x))
```

#### 1.2. Count plot
- A special case of Bar plot

#### 1.3 Point plot

- Either use catplot with kind='point' or directly use pointplot()


```python
sns.catplot(y="total_cases", x="location", kind="point", data=w_df)
```

```python
sns.catplot(y="new_cases", x="location", hue='Year',  kind="point", data=w_df)
```

## 3 Observing distribution with different categories Using cat plot

#### When there more and more data flows in,  categorical scatter plots becomes infeasible hence trying to get distrbution related to each category would help us summarize the data.

####  there are several approaches for summarizing the distributional information in ways that facilitate easy comparisons across the category levels.corresponding to the categorical variable

- boxplot() (in catplot kind="box")

- violinplot() (in catplot kind="violin")

- boxenplot() (in catplot kind="boxen")

### 3.1 Boxplots

- Boxplot is a method for graphically depicting groups of numerical data through their quartiles. Box plots may also have lines extending from the boxes (whiskers) indicating variability outside the upper and lower quartiles.


- The “whiskers” extend to points that lie within 1.5 IQRs of the lower and upper quartile and other outliers are displayed independently.

[For further parameters related with box plot](https://seaborn.pydata.org/generated/seaborn.boxplot.html#seaborn.boxplot)


```python
sns.catplot(x="location", hue = 'Year',y="new_cases", kind="box", data=w_df, height=9)
```

### 3.2 Boxen plot

#### Almost same functionality of Box plot but it gives slightly more information about the distrbution than box plots. Mostly preferred when the data is ver large

[boxenplot](https://seaborn.pydata.org/generated/seaborn.boxenplot.html#seaborn.boxenplot)

- Look for k-depth parameter if you want fine tune the boxes

```python
sns.catplot(x="location", hue = 'Year',y="new_cases", kind="boxen", data=w_df, height=9)
```

### 3.3 Violin Plot

#### A different approach to understand the distributions of different categories over a specific numeric variable. This plot tries to combine the box plot with kernel density estimation procedure to get a smooth graphs.

-  Additionally, the quartile and whisker values from the boxplot are shown inside the violin. 
-  Since we are using KDE there are some additional parameter that can be used to finetune the estimation process.

[For further parameters in violin plot](https://seaborn.pydata.org/generated/seaborn.violinplot.html#seaborn.violinplot)


```python
sns.catplot(x="location", hue = 'Year',y="new_cases", kind="violin", data=w_df, height=9)
```

```python
sns.catplot(x="location", hue = 'Year',y="new_cases", kind="violin", data=w_df, height=9, bw=0.1)
```

## Functions to draw linear regression models

#### There are 2 plots that is used for this purpose. 
1. regplot() - Accept date in any format like numy series, pandas series or pandas dataframe and string columns [For other parameters](https://seaborn.pydata.org/generated/seaborn.regplot.html)
2. lmplot() - Only accept data from padas dataframe where the variables will be specifed using string notation [For other parameters](https://seaborn.pydata.org/generated/seaborn.lmplot.html)

- Plot data and a linear regression model fit.
- In the simplest invocation, both functions draw a scatterplot of two variables, x and y, and then fit the regression model y ~ x and plot the resulting regression line and a 95% confidence interval for that regression

[Please follow this tutorial for additional information](https://seaborn.pydata.org/tutorial/regression.html#regression-tutorial)

```python
sns.set_style('whitegrid')
reg_ = sns.regplot(x='new_cases', y='new_deaths', data = w_df,
            scatter_kws = {'color': 'Blue', 'alpha':0.3},
            line_kws = {'color': 'red', 'alpha':0.4})
reg_.figure.set_size_inches(10, 10)
```

```python
fig, (ax1) = plt.subplots(ncols=1)
sns.regplot(data=w_df,y='new_cases', x='total_cases', ax=ax1, label='Total vs new cases')
ax2 = plt.twinx()
sns.regplot(data=w_df,y='new_deaths', x='total_cases', ax=ax2, color = 'red',label = 'Total vs new deaths',scatter_kws={'alpha':0.3})
ax1.figure.legend()
```

```python
sns.set_style('whitegrid')
sns.lmplot(x='new_cases', y='new_deaths', data = w_df, height = 10, hue = 'Year', col='location', col_wrap=2)
```

### Joint plot

- This is one of the widely used grpah to understand the distribution of the data.
- Main advantage is it helps us to analyse both bivariate data and univariet data in one plot.

[Joint plot Additional parameters](http://seaborn.pydata.org/generated/seaborn.jointplot.html)

```python
sns.set()
sns.jointplot(data=w_df, x='new_cases', y='new_deaths')
```

- We can specify a plot kind and the jointplot automatically define other variables and plot. We can use plots scatter, regression, kde, histogram, hex, and residual.

```python

sns.jointplot(data=w_df, x='new_cases', y='new_deaths', kind='reg')
sns.jointplot(data=w_df, x='new_cases', y='new_deaths', kind='hex',height=15)
```

### Directly plotting residuals

```python
sns.jointplot(data=w_df, x='new_cases', y='new_deaths' ,kind='resid');
```

```python
sns.jointplot(data=w_df, x='new_cases', y='new_deaths', hue='Year');
```

### Customing the graph using Joint Grid

```python
joint_ex = sns.JointGrid(data=w_df, x='new_cases', y='new_deaths')
joint_ex.plot(sns.scatterplot, sns.violinplot); ## first plot for bivariate and second plot for univariate 
```

```python
joint_ex = sns.JointGrid(data=w_df, x='new_cases', y='new_deaths')
joint_ex.plot(sns.kdeplot, sns.boxenplot);
```
