```python
import pandas as pd
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as shc
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

# Generating few clusters with random data

```python
data,cluster=make_blobs(n_samples=50,centers=3)
```

# Creating scipy linkage object

```python
result=shc.linkage(data, method='ward')
```

# Dendrogram

```python
plt.figure(figsize=(10,10))
shc.dendrogram(result)
plt.show()
```

# Creating sklearn Agglomerative Clustering object 

```python
acl=AgglomerativeClustering(n_clusters=3,affinity="euclidean",linkage="ward")
```

# Fitting the model to data

```python
acl.fit(data)
```

# Clusters

```python
acl.labels_
```
